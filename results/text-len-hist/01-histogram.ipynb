# Distribution of tweet lengths

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
mpl.style.use('bmh')
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('counts.csv')
```

```python
df200 = df[df['text_len'] < 150]
```

```python
plt.figure(figsize=(15, 5))

plt.hist(
    df200['text_len'],
    weights=df200['count'],
    bins=df200['text_len'].max(),
)

plt.title('Tweet character counts, ~100M tweets from June 2017')
plt.xlabel('Tweet length')
plt.ylabel('Character count')
plt.show()
```
